```python
import numpy as np 
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import myfun as wf
import math, gc, logging,os
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LongitudeFormatter, LatitudeFormatter
import geocat.viz as gv
```

```python
if __name__ == "__main__":
    fpath_in0 = "/glade/derecho/scratch/hcluo/VerticalProfile/5NS/"
    fpath_in1 = "/glade/derecho/scratch/hcluo/VerticalProfile/5_15N/"
    years = np.arange(1997,2024)
    fnames1_1 = []
    fnames2_1 = []
    fnames3_1 = []
    fnames4_1 = []
    fnames5_1 = []
    fnames6_1 = []
    fnames7_1 = []
    fnames8_1 = []
    fnames1_2 = []
    fnames2_2 = []
    fnames3_2 = []
    fnames4_2 = []
    fnames5_2 = []
    fnames6_2 = []
    fnames7_2 = []
    fnames8_2 = []
    for i in years:
        fnames1_1.append(fpath_in0+"v_"+str(i)+".nc")
        fnames2_1.append(fpath_in0+"MSE_"+str(i)+".nc")
        fnames3_1.append(fpath_in0+"vMSE_"+str(i)+".nc")
        fnames4_1.append(fpath_in0+"vMSE_transient_"+str(i)+".nc")
        fnames5_1.append(fpath_in0+"v_anno2_"+str(i)+".nc")
        fnames6_1.append(fpath_in0+"MSE_anno2_"+str(i)+".nc")
        fnames7_1.append(fpath_in0+"Lvq_anno2_"+str(i)+".nc")
        fnames8_1.append(fpath_in0+"Lvq_"+str(i)+".nc")
        fnames1_2.append(fpath_in1+"v_"+str(i)+".nc")
        fnames2_2.append(fpath_in1+"MSE_"+str(i)+".nc")
        fnames3_2.append(fpath_in1+"vMSE_"+str(i)+".nc")
        fnames4_2.append(fpath_in1+"vMSE_transient_"+str(i)+".nc")
        fnames5_2.append(fpath_in1+"v_anno2_"+str(i)+".nc")
        fnames6_2.append(fpath_in1+"MSE_anno2_"+str(i)+".nc")
        fnames7_2.append(fpath_in1+"Lvq_anno2_"+str(i)+".nc")
        fnames8_2.append(fpath_in1+"Lvq_"+str(i)+".nc")
        
```

```python
ds1_1 = xr.open_mfdataset(fnames1_1,concat_dim = 'time', combine='nested')["v_vertical"].mean("time").sel(level = slice(00,1000))
ds1_2 = xr.open_mfdataset(fnames1_2,concat_dim = 'time', combine='nested')["v_vertical"].mean("time").sel(level = slice(00,1000))
plt.plot(ds1_1,ds1_1.level,"dodgerblue",label = r"$\overline{[v]}|_{5S}^{5N}$")
plt.plot(ds1_2,ds1_2.level,"darkorange",label = r"$\overline{[v]}|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
ds2_1 = xr.open_mfdataset(fnames2_1,concat_dim = 'time', combine='nested')["MSE_vertical"].mean("time").sel(level = slice(200,1000))
ds2_2 = xr.open_mfdataset(fnames2_2,concat_dim = 'time', combine='nested')["MSE_vertical"].mean("time").sel(level = slice(200,1000))
plt.plot(ds2_1,ds2_1.level,"dodgerblue",label = r"$\overline{[h]}|_{5S}^{5N}$")
plt.plot(ds2_2,ds2_2.level,"darkorange",label = r"$\overline{[h]}|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
ds3_1 = xr.open_mfdataset(fnames3_1,concat_dim = 'time', combine='nested')["vMSE_full_vertical"].mean("time").sel(level = slice(00,1000))
ds3_2 = xr.open_mfdataset(fnames3_2,concat_dim = 'time', combine='nested')["vMSE_full_vertical"].mean("time").sel(level = slice(00,1000))
plt.plot(ds3_1,ds3_1.level,"dodgerblue",label = r"$\overline{[vh]}|_{5S}^{5N}$")
plt.plot(ds3_2,ds3_2.level,"darkorange",label = r"$\overline{[vh]}|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
ds4_1 = xr.open_mfdataset(fnames4_1,concat_dim = 'time', combine='nested')["vMSE_transient_vertical"].mean("time").sel(level = slice(00,1000))
ds4_2 = xr.open_mfdataset(fnames4_2,concat_dim = 'time', combine='nested')["vMSE_transient_vertical"].mean("time").sel(level = slice(00,1000))
plt.plot(ds4_1,ds4_1.level,"dodgerblue",label = r"$\overline{[v^*'h^*']}|_{5S}^{5N}$")
plt.plot(ds4_2,ds4_2.level,"darkorange",label = r"$\overline{[v^*'h^*']}|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
ds5_1 = xr.open_mfdataset(fnames5_1,concat_dim = 'time', combine='nested')["v_anno2_vertical"].mean("time").sel(level = slice(0,1000))
ds5_2 = xr.open_mfdataset(fnames5_2,concat_dim = 'time', combine='nested')["v_anno2_vertical"].mean("time").sel(level = slice(0,1000))
plt.plot(ds5_1,ds5_1.level,"dodgerblue",label = r"$\overline{[v^*'^2]}|_{5S}^{5N}$")
plt.plot(ds5_2,ds5_2.level,"darkorange",label = r"$\overline{[v^*'^2]}|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
ds6_1 = xr.open_mfdataset(fnames6_1,concat_dim = 'time', combine='nested')["MSE_anno2_vertical"].mean("time").sel(level = slice(0,1000))
ds6_2 = xr.open_mfdataset(fnames6_2,concat_dim = 'time', combine='nested')["MSE_anno2_vertical"].mean("time").sel(level = slice(0,1000))
ds7_1 = xr.open_mfdataset(fnames7_1,concat_dim = 'time', combine='nested')["Lvq_anno2_vertical"].mean("time").sel(level = slice(0,1000))
ds7_2 = xr.open_mfdataset(fnames7_2,concat_dim = 'time', combine='nested')["Lvq_anno2_vertical"].mean("time").sel(level = slice(0,1000))
Lv = 2.5e6
plt.plot(ds6_1,ds6_1.level,"dodgerblue",label = r"$MSE|_{5S}^{5N}$")
plt.plot(ds6_2,ds6_2.level,"darkorange",label = r"$MSE|_{5N}^{15N}$")
plt.plot(Lv*ds7_1,ds7_1.level,ls = ":",color = "dodgerblue",label = r"$L_vq|_{5S}^{5N}$")
plt.plot(Lv*ds7_2,ds7_2.level,ls = ":",color = "darkorange",label = r"$L_vq|_{5N}^{15N}$")
plt.gca().invert_yaxis()
plt.legend()
plt.grid(linestyle = ':', linewidth = 0.5)
```

```python
fig = plt.figure(figsize=(7, 7),layout='compressed')
    
gs = fig.add_gridspec(2, 2,wspace=0.02, hspace=0.05)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])
ax1.plot(ds3_1/1.0e5,ds3_1.level,"dodgerblue",label = "5S-5N")
ax1.plot(ds3_2/1.0e5,ds3_2.level,"darkorange",label = "5N-15N")
ax2.plot(ds4_1/1.0e3,ds4_1.level,"dodgerblue",label = "5S-5N")
ax2.plot(ds4_2/1.0e3,ds4_2.level,"darkorange",label = "5N-15N")
ax3.plot(ds5_1,ds5_1.level,"dodgerblue",label = "5S-5N")
ax3.plot(ds5_2,ds5_2.level,"darkorange",label = "5N-15N")
ax4.plot(ds6_1/1.0e7,ds6_1.level,"dodgerblue",label = r"$h|_{5S}^{5N}$")
ax4.plot(ds6_2/1.0e7,ds6_2.level,"darkorange",label = r"$h|_{5N}^{15N}$")
ax4.plot(Lv*ds7_1/1.0e7,ds7_1.level,ls = ":",color = "dodgerblue",label = r"$L_vq|_{5S}^{5N}$")
ax4.plot(Lv*ds7_2/1.0e7,ds7_2.level,ls = ":",color = "darkorange",label = r"$L_vq|_{5N}^{15N}$")

ax1.set_ylabel("Pressure (hPa)")
ax3.set_ylabel("Pressure (hPa)")
subtitles = ["a)","b)","c)","d)"]
xlabels = [r"$\overline{[vh]}\;(\times 10^5\;W \,m \,kg^{-1})$",
           r"$\overline{[v^*'h^*']}\;(\times 10^3\;W \,m \,kg^{-1})$",
           r"$\overline{[v^*'^2]}\;(m^2 \,s^{-2})$",
           r"$\overline{[h^*'^2]} & \overline{[(L_vq)^*'^2]}\;(\times 10^7\;J^2 \,kg^{-2})$",]
axs = [ax1,ax2,ax3,ax4]
for i in range(len(axs)):
    ax = axs[i]
    ax.set_title(xlabels[i])
    ax.text(0.03, 0.97, subtitles[i], fontsize = 13,transform=ax.transAxes,va='top', ha='left')
    ax.invert_yaxis()
    ax.legend(loc = "upper right")
    ax.grid(linestyle = ':', linewidth = 0.5)
plt.savefig("VerticalProfile.pdf")
```
